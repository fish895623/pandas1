```python
import networkx as nx


G = nx.Graph()
G
```

```python
G.add_node(1)
print(G.nodes())
```

```python
G.add_node('P')
G.add_node('HI')
print(G.nodes())
```

```python
G.add_nodes_from([2, 3])
print(G.nodes())
```

```python
nx.draw(G, with_labels=True)
```

```python
G.add_edge(1,2)
nx.draw(G, with_labels=True, edge_color='grey')
```

```python
print(G.edges())
```

```python
G = nx.path_graph(4)
nx.add_path(G, [0, 1, 2, 3])
nx.draw(G, with_labels=True)
```

```python
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5])
G.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (4, 5)])
nx.draw(G, with_labels=True)
```

```python
G.remove_node(3)
nx.draw(G, with_labels=True)
```
