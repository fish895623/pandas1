```python
# 값들중 가장 큰값 구하기
# 큰값을 가진 주인은?
import numpy as np
import networkx as nx

G = nx.karate_club_graph()
degree = [d for n, d in G.degree()]
nx.draw(G, with_labels=True)
a = G.degree()
a = dict(a)

ddd = {v: k for k, v in a.items()}
print("가장큰애")
print(ddd[max(ddd.keys())])
print("그의 값")
print(max(ddd))
```
