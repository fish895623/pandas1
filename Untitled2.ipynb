```python
labels = 'a', 'b', 'c', 'd'
size = [12, 20, 15, 3]
fig, ax = plt.subplots()
ax.pie(size, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(size, labels=labels, autopct='%.1f%%', startangle=90)
plt.show()
```

autopct 자동으로 퍼센트로 바꾸어서 나타내어 준다

```python
sizes=[6,10,15,9]
# a, b, c, d  
explode=(0,0.1,0.1,0)

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, shadow=True,autopct='%.1f%%', startangle=90)
plt.show()
```
